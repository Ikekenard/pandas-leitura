```python
import json
import requests
```

```python
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
resutados = json.loads(dados_frutas.text)
```

```python
dados_frutas_normalizado = pd.json_normalize(resutados)
dados_frutas_normalizado.head()
```
